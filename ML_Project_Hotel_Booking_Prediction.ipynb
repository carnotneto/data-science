```python
def week_function(row):
    feature_1 = 'stays_in_weekend_nights'
    feature_2 = 'stays_in_week_nights'
    
    if row[feature_2] == 0 and row[feature_1] > 0:
        return 'stays_only_for_the_weekend'
    elif row[feature_1] == 0 and row[feature_2] > 0:
        return 'stays_only_weekdays'
    elif row[feature_2] > 0 and row[feature_1] > 0:
        return 'stays_both_weekends_and_weekdays'
    else:
        return 'undefined_data'
```

```python
data_2=bookings[bookings['is_canceled']==0]
data_2['weekend_or_weekdays']=data_2.apply(week_function, axis=1)
```

Applying Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
logreg = LogisticRegression(solver='liblinear')
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

Applying Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(logreg, x, y, cv=10)
score
```

Applying Multiple Machine Learn Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models = []

models.append(('Logistic Regression', LogisticRegression(solver='liblinear')))
models.append(('Naive Bayes', GaussianNB()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('Random Forest', RandomForestClassifier()))
models.append(('Decision Tree', DecisionTreeClassifier()))
```

```python
models
```

```python
for name,model in models:
    print(name)
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    
    print('Confusion Matrix')  
    print(confusion_matrix(predictions,y_test))
    print(f'accuracy = {accuracy_score(predictions,y_test)}')
    print('\n')
```
